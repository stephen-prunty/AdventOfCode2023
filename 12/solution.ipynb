```python
import itertools as it

def get_combinations(n,p) :
    return list(it.combinations(range(n),p))

def get_layout(springs): 
    current_sum = 0
    number_layout = []
    for i in range(len(springs)) :
        if springs[i] == '#' :
            current_sum += 1
        else :
            if current_sum != 0 :
                number_layout.append(current_sum)
                current_sum = 0
    if current_sum != 0 :
        number_layout.append(current_sum)

    return number_layout

def get_q_indicies(springs) :
    q_indicies = []
    for i in range(len(springs)) :
        if springs[i] == '?' :
            q_indicies.append(i)
    return q_indicies

def no_of_missing_springs(springs,layout) :
    return sum(layout) - springs.count('#')

def sum_arrangements(springs,layout) :
    q_indicies = get_q_indicies(springs)
    missing_springs = no_of_missing_springs(springs,layout)
    combinations = get_combinations(len(q_indicies),missing_springs)
    sum = 0

    for combination in combinations :
        new_springs = springs
        for i in combination :
            new_springs= new_springs[:q_indicies[i]] + '#' + new_springs[q_indicies[i]+1:]
        if get_layout(new_springs) == layout :
            sum += 1
    
    return sum

# Part 2

def find_no_of_q_at_start(springs) :
    for i in range(len(springs)) :
        if springs[i] == '.' :
            return i
        if springs[i] == '#' :
            return i - 1
    return len(springs) + 1

def find_no_of_q_at_end(springs) :
    for i in range(len(springs)) :
        if springs[-(i+1)] == '.' :
            return i
        if springs[-(i+1)] == '#' :
            return i - 1
    return len(springs) + 1

def sum_arrangements_n(springs,layout,n) :
    if springs[0] == '#' :
        left_springs = springs
    else :
        start_springs = '?' * find_no_of_q_at_start(springs)
        left_springs = springs + '?' + start_springs
    
    if springs[-1] == '#' :
        right_springs = springs
    else :
        end_springs = '?' * find_no_of_q_at_end(springs)
        right_springs = end_springs + '?' + springs

    maximum = max((sum_arrangements(left_springs,layout),sum_arrangements(right_springs,layout)))
    
    return maximum**(n-1) * sum_arrangements(springs,layout)
```
