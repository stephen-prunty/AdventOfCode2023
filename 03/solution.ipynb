```python
# Part 1

file_data = open('input.txt','r')
lines = file_data.readlines()

# go through each number and see if it's adjacent to a symbol.

star_numbers = [[0 for col in range(140)] for row in range(140)]
star_ratios = [[1 for col in range(140)] for row in range(140)]


def find_sum(grid) :

    sum = 0
    
    for line in grid :

        #print("Line #" + str(grid.index(line)) + ":")

        for i in range(len(line)) :
            num = 0
            proceed = False

            if line[i].isnumeric() :

                if i == 0 :
                    proceed = True
                else :
                    if not line[i-1].isnumeric() :
                        proceed = True
            
            if proceed :
                j = i + 1
                x = line[i:j]
                while line[i:j].isnumeric() :
                    j = j+1
                num = int(line[i:j-1])
                

                if is_adj_char_symbol(i,j-1,grid.index(line),grid) : 
                    #print(num)
                    sum += num
        
        #print("Current Sum: " + str(sum))
        #print()

    return sum
                


def is_symbol(char) :
    return not (char.isnumeric() or char == '.' or char == ' ' or char == '') 

def is_star(char) :
    return char == '*'


def is_adj_char_symbol(start_x, end_x, line_no, grid):
    symbol_found = False

    min_x = max(start_x - 1, 0)
    max_x = min(end_x + 1, len(grid[line_no]) - 1)

    #print(len(grid[line_no]))

    min_y = max(line_no - 1, 0)
    max_y = min(line_no + 2 , len(grid))

    #print(min_x)
    #print(max_x)

    #print(min_y)
    #print(max_y)


    for y in range(min_y,max_y) :
        for x in range (min_x, max_x) :
            #print(grid[y][x])
            if is_symbol(grid[y][x]) :
                symbol_found = True
                if is_star(grid[y][x]) :
                    star_numbers[y][x] += 1
                    star_ratios[y][x] *= int(grid[line_no][start_x:end_x])
    
    return symbol_found


#is_adj_char_symbol(137,140,21,lines)

find_sum(lines)

#print(is_symbol('%'))
for line in star_ratios :
    print(line)

```
