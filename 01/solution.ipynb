```python
# Part 1
calibrations = open('input.txt','r').readlines()

def find_sum(array):
    sum = 0
    for line in array :
        first_found = False
        first_digit = 0
        last_digit = 0

        for char in line :
            
            if char.isdigit() :

                if (not first_found) : 
                    first_digit = int(char)

                    first_found = True

                last_digit = int(char)
                
        
        sum += 10*first_digit + last_digit

    return sum



find_sum(calibrations)
```

```python
# Part 2

calibrations = open('input.txt','r').readlines()
calibrations = list(map(lambda x:x.strip(),calibrations))


def replace_word_digits(text) :
    text = replace_first_digit(text)
    text = replace_last_digit(text)
    return text

def replace_first_digit(text):
    digit_words = {
        'zero': '0',
        'one': '1',
        'two': '2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9'
    }
    
    found = False
    new_string = text

    for i in range(len(text)) :
        for word,digit in digit_words.items() :
            if text[i].isnumeric() : found = True
            if text[:i].find(word) >= 0 :
                new_string = text[:i] + digit + text[i:]
                found = True
        if found : break
    
    return new_string

def replace_last_digit(text):
    digit_words = {
        'zero': '0',
        'one': '1',
        'two': '2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9'
    }
    
    found = False
    new_string = text

    for i in reversed(range(len(text))) :
        for word,digit in digit_words.items() :
            if text[i].isnumeric() : found = True
            if text[i:].find(word) >= 0 :
                new_string = text[:i] + digit + text[i:]
                found = True
        if found : break
    
    return new_string


def find_sum_2(array):
    sum = 0
    # print("sum = 0")
    # print()
    for line in array :
        # print("Line: " + str(line))
        # print("Value: " + str(find_line_value(line)))
        # print("Reformatted: " + str(replace_word_digits(line)))
        # print("Reformatted Value: " + str(find_line_value(replace_word_digits(line))))
        # print("Sum = " + str(sum) + " + " + str(find_line_value(replace_word_digits(line))) + " = " + str(sum + find_line_value(replace_word_digits(line))))
        # print()

        sum += find_line_value(replace_word_digits(line))

    return sum

def find_line_value(line) :
    first_found = False
    first_digit = 0
    last_digit = 0
    for char in line :
            
            if char.isdigit() :

                if (not first_found) : 
                    first_digit = int(char)

                    first_found = True

                last_digit = int(char)
                
    return 10*first_digit + last_digit



find_sum_2(calibrations)

# test = 'eight83oneight9'

# replace_word_digits(test)
```
