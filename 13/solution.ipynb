```python
# Main Code:

import copy

def data_parser(input_file) :
    file_data = open(input_file,'r').readlines()
    lines = [line.strip() for line in file_data]

    patterns = []
    current_pattern = []

    for line in lines :
        
        if line != '' :
            current_pattern.append(list(line))
        else :
            patterns.append(current_pattern)
            current_pattern = []

    patterns.append(current_pattern)

    return patterns


def row_ref_index(pattern, ignore=None) :

    for i in range(len(pattern) - 1) :
        #print(f"--- Checking between row {i} & {i+1} ---")
        range_to_check = min(i + 1, len(pattern) - 1 - i)
        check = True

        if i == ignore :
            check = False

        for j in range(range_to_check) :
            #print("Comparing row " + str(i - j) + " & " + str(i + j + 1) + " : " + pattern[i - j] + " & " + pattern[i + j + 1])
            if pattern[i - j] != pattern[i + j + 1] :
                check = False
                break
        
        if check :
            return i
    
    return -1


def transpose(pattern) :
    return list(map(list, zip(*pattern)))

def print_pattern(pattern) :
    for row in pattern :
        for col in row :
            print(col, end = '')
        print()
    print()

def fix_smudge(pattern, row_no, col_no) :
    return_pattern = copy.deepcopy(pattern)
    
    if pattern[row_no][col_no] == '.' :
        return_pattern[row_no][col_no] = '#'
    else :
        return_pattern[row_no][col_no] = '.'
    
    return return_pattern
```
